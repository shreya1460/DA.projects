#### Q2. In which year highest number of the TV Shows & Movies were released? Show with Bar Graph.

dtypes

```python
data.dtypes
```

to_datetime

Changing Release_Date data type into to_datetime

```python
data["Date_N"] = pd.to_datetime(data["Release_Date"])
```

```python
data.head(2)
```

```python
data.dtypes
```

#### dt.year.value_counts()

```python
data["Date_N"].dt.year.value_counts()       #It counts the occurrence of all individual Years in Date
```

```python
ax = data["Date_N"].dt.year.value_counts().plot(kind = "bar")        #To showing the data in which year highest Tv show and movies released
for container in ax.containers:
    ax.bar_label(container)
```

#### Q3. How many Movies & TV Shows are in the dataset? Show with Bar Graph.

```python
data["Category"].value_counts().plot(kind="bar")            #To group all unique items of a column and show their count by using value_count method.     
```

```python
data.groupby("Category").Category.count()           #To group all unique items of a column and show their count by using group by method.
```

```python
ax = data["Category"].value_counts().plot.bar(color = ["b","g"])

for i in ax.containers:
    ax.bar_label(i)
plt.show()
```
